```python
# # Step 1: Read the required packages from requirements.txt
# required_packages = []
# with open('requirements.txt', 'r') as file:
#     required_packages = [line.strip() for line in file if line.strip() and not line.startswith('#')]

# # Step 2: Get the list of installed packages
# import pkg_resources
# installed_packages = {pkg.key: pkg.version for pkg in pkg_resources.working_set}

# # Step 3: Check if all required packages are installed
# missing_packages = []
# for requirement in required_packages:
#     # Split the requirement to get the package name
#     package_name = requirement.split('==')[0] if '==' in requirement else requirement
#     if package_name not in installed_packages:
#         missing_packages.append(package_name)

# # Output the result
# if not missing_packages:
#     print("All packages are successfully installed.")
# else:
#     print(f"Missing packages: {missing_packages}")


import subprocess

# Step 1: Read the required packages from requirements.txt
required_packages = []
with open('requirements.txt', 'r') as file:
    required_packages = [line.strip() for line in file if line.strip() and not line.startswith('#')]

# Step 2: Get the list of installed packages
import pkg_resources
installed_packages = {pkg.key: pkg.version for pkg in pkg_resources.working_set}

# Step 3: Check if all required packages are installed
missing_packages = []
for requirement in required_packages:
    # Split the requirement to get the package name and version
    if '==' in requirement:
        package_name, package_version = requirement.split('==')
    else:
        package_name, package_version = requirement, None

    # Check if the package is installed
    if package_name not in installed_packages:
        missing_packages.append(requirement)
    elif package_version and installed_packages[package_name] != package_version:
        missing_packages.append(requirement)  # Version mismatch

# Step 4: Install missing packages
if not missing_packages:
    print("All packages are successfully installed.")
else:
    print(f"Missing packages: {missing_packages}")

    # Install missing packages
    for package in missing_packages:
        print(f"Installing: {package}")
        subprocess.check_call([f"pip install {package}"], shell=True)

    print("Missing packages installed successfully.")

```
